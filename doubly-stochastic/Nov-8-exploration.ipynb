```julia
import ImageView: imshow
import Images: Gray
import TestImages: testimage
```

```julia
mandrill = 1*Gray.(testimage("mandrill"))
```

```julia
m = convert(Matrix{Float32}, mandrill)
```

```julia
function normalize_image(im)
    im_res = im .- minimum(im)
    return (1/maximum(im_res))*im_res
end
```

```julia
import LinearAlgebra: transpose, norm
```

```julia
t = [1 2; 3 4]
```

```julia
(x->x+1).(t)
```

```julia
sum(t, dims=1)
```

```julia
sum(t, dims=2)
```

```julia
m_plus = (x->x+1).(m)
```

```julia
sum(m, dims=1)
```

```julia
m_norm_col1 = m_plus ./ sum(m_plus, dims=1)
```

```julia
sum(m_norm_col1, dims=2)
```

```julia
sum(m_norm_col1, dims=1)
```

```julia
m_norm_row2 = m_norm_col1 ./ sum(m_norm_col1, dims=2)
```

```julia
sum(m_norm_row2, dims=1)
```

```julia
m_norm_col3 = m_norm_row2 ./ sum(m_norm_row2, dims=1)
```

```julia
sum(m_norm_col3, dims=2)
```

```julia
m_norm_row4 = m_norm_col3 ./ sum(m_norm_col3, dims=2)
```

```julia
sum(m_norm_row4, dims=1)
```

```julia
sum(m_norm_row4, dims=2)
```

```julia
mdc = m_norm_row4
```

```julia
Gray.(normalize_image(transpose(mdc)*mdc))
```

```julia
Gray.(normalize_image(mdc*mdc))
```

```julia
Gray.(normalize_image(transpose(mdc)*transpose(mdc)))
```
